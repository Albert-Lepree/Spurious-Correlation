```python
from db import query
import pandas as pd

# Pull bonferroni results as proxy labels for meta-feature regression
# Rejected = spurious, Not rejected = genuine
bonf = query("SELECT feature, feature_group, bonferroni_rejected FROM bonferroni_fdr_results")
meta = pd.read_csv('results/metafeature_descriptors.csv')

# For labeled features only (control + spurious), check agreement with bonferroni
labeled = bonf[bonf['feature_group'].isin(['control','spurious'])].copy()
labeled['detection_label'] = labeled['bonferroni_rejected'].astype(int)

print("Original labels vs Bonferroni detection:")
print(labeled.groupby(['feature_group','detection_label']).size())
```

```python
labeled['true_label'] = (labeled['feature_group'] == 'spurious').astype(int)
disagree = labeled[labeled['true_label'] != labeled['detection_label']]
display(disagree[['feature','feature_group','true_label','detection_label']])
```

```python
# This becomes the training set for 03_meta_feature_regression.ipynb
labeled.to_csv('Analysis/labeled_features.csv', index=False)
print(f"Saved {len(labeled)} labeled features")
```
